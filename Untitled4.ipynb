```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

Classical Song:

```python
# Define file path
file_path = "Piya Tose Naina Laage Re - Guide (1965).mp3"  # Change if needed

# Load audio file
y, sr = librosa.load(file_path, sr=None)  # sr=None keeps the original sample rate

# Compute Short-Time Fourier Transform (STFT)
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

# Plot the Spectrogram
plt.figure(figsize=(12, 6))
librosa.display.specshow(D, sr=sr, x_axis="time", y_axis="log")
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram of 'Piya Tose Naina Laage Re'")
plt.xlabel("Time (seconds)")
plt.ylabel("Frequency (Hz)")
plt.show()
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
from scipy.signal import get_window

# Load audio file (replace with your actual file path)
file_path = '98223-7-0-0.wav'
y, sr = librosa.load(file_path, sr=None)

# Function to compute and plot spectrogram with different windows
def plot_spectrogram(y, sr, window_type, window_length):
    # Use librosa's stft with the desired window
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y, n_fft=window_length, hop_length=window_length//2, window=window_type)), ref=np.max)

    plt.figure(figsize=(12, 6))
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format="%+2.0f dB")
    plt.title(f'Spectrogram using {window_type} Window')
    plt.show()

# Generate and plot spectrograms using different windows
plot_spectrogram(y, sr, 'hann', 2048)  # Hann Window
plot_spectrogram(y, sr, 'hamming', 2048)  # Hamming Window
plot_spectrogram(y, sr, 'boxcar', 2048)  # Rectangular Window (Boxcar)
```

# **Hann Window:**

**Appearance**: Smooth transitions with sharp, well-defined frequency peaks.

**Leakage**: Minimal leakage, making it clear and clean.

**Frequency Resolution**: Good, with distinct frequency components.

# **Hamming Window:**

**Appearance**: Similar to Hann, but with slightly smoother peaks.

**Leakage**: Even less leakage than Hann.

**Frequency Resolution**: Slightly reduced compared to Hann, but clearer overall.

# **Rectangular Window:**

**Appearance**: Blurry peaks with wide energy spread.

**Leakage**: High leakage, making the spectrogram appear fuzzy.

**Frequency Resolution**: Higher resolution but less precise due to leakage.

```python
import os
import zipfile
import requests
import librosa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import librosa.display
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Define Dataset URL and Paths
dataset_url = "https://zenodo.org/record/1203745/files/UrbanSound8K.tar.gz"
dataset_folder = "UrbanSound8K"
dataset_zip = "UrbanSound8K.tar.gz"

# Step 1: Download Dataset if Not Already Present
if not os.path.exists(dataset_folder):
    print("Downloading UrbanSound8K dataset...")
    response = requests.get(dataset_url, stream=True)
    with open(dataset_zip, "wb") as file:
        for chunk in response.iter_content(chunk_size=1024):
            file.write(chunk)
    print("Download completed.")

    # Step 2: Extract Dataset
    print("Extracting dataset...")
    os.system(f"tar -xzf {dataset_zip}")
    print("Extraction completed.")

# Step 3: Load Metadata
metadata_file = os.path.join(dataset_folder, "metadata/UrbanSound8K.csv")
metadata = pd.read_csv(metadata_file)

# Function to Extract MFCC Features
def extract_features(file_path, window_type='hann', window_length=2048):
    y, sr = librosa.load(file_path, sr=None)

    # Get window function
    window = librosa.filters.get_window(window_type, window_length)

    # Compute STFT with selected window
    D = librosa.stft(y, n_fft=window_length, window=window)
    D_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

    # Extract MFCCs
    mfccs = librosa.feature.mfcc(S=D_db, sr=sr, n_mfcc=13)
    return np.mean(mfccs, axis=1)

# Function to Load Data from Audio Files
def load_data(window_type='hann'):
    features, labels = [], []

    for _, row in metadata.iterrows():
        file_path = os.path.join(dataset_folder, f"audio/fold{row['fold']}/{row['slice_file_name']}")

        if os.path.exists(file_path):
            feature_vector = extract_features(file_path, window_type)
            features.append(feature_vector)
            labels.append(row["class"])

    return np.array(features), np.array(labels)

# Function to Train SVM Classifier
def train_svm_classifier(features, labels):
    label_encoder = LabelEncoder()
    labels_encoded = label_encoder.fit_transform(labels)

    # Split dataset
    X_train, X_test, y_train, y_test = train_test_split(features, labels_encoded, test_size=0.2, random_state=42)

    # Train SVM
    clf = svm.SVC(kernel='linear')
    clf.fit(X_train, y_train)

    # Evaluate Accuracy
    y_pred = clf.predict(X_test)
    return accuracy_score(y_test, y_pred)

# Evaluate Different Windowing Techniques
window_types = ['hann', 'hamming', 'boxcar']
accuracies = {}

for window_type in window_types:
    print(f"\nTraining SVM with {window_type} window...")
    features, labels = load_data(window_type)
    accuracy = train_svm_classifier(features, labels)
    accuracies[window_type] = accuracy
    print(f"Accuracy with {window_type} window: {accuracy:.4f}")

# Print Final Comparison
print("\nFinal Accuracy Comparison:")
for window, acc in accuracies.items():
    print(f"{window.capitalize()} Window: {acc:.4f}")
```
